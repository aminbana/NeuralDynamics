```python
def visualise_connectivity(S):
    Ns = len(S.source)
    Nt = len(S.target)
    figure(figsize=(10, 4))
    subplot(121)
    plot(zeros(Ns), arange(Ns), 'ok', ms=10)
    plot(ones(Nt), arange(Nt), 'ok', ms=10)
    for i, j in zip(S.i, S.j):
        plot([0, 1], [i, j], '-k')
    xticks([0, 1], ['Source', 'Target'])
    ylabel('Neuron index')
    xlim(-0.1, 1.1)
    ylim(-1, max(Ns, Nt))
    subplot(122)
    plot(S.i, S.j, 'ok')
    xlim(-1, Ns)
    ylim(-1, Nt)
    xlabel('Source neuron index')
    ylabel('Target neuron index')
```

```python
start_scope()

IeRm = 25*mV
taum = 20*ms
El = -70*mV
Vth = -54*mV
rm = 10 * Mohm
#  - gs*(V-Es)*rm
eqs = '''
dv/dt = (IeRm - (v-El))/taum : volt
'''

G = NeuronGroup(2, eqs,threshold='v > Vth', reset='v = El - 10*mV', method='exponential_euler')
M = StateMonitor(G, ['v'], record=True)
spikemon = SpikeMonitor(G)

G.v = 0



run(1000*ms)


figure()

plot(M.t/ms, M.v[0]/mV, c='black')
for t in spikemon.t:
    axvline(t/ms, c='black')
xlabel('Time (ms)')
ylabel('v');

# plot(M.t/ms, M.gs[0]/nsiemens)
xlabel('Time (ms)')
ylabel('v');
ylim([-80,0])
xlim([0,100])

freq = np.sum (spikemon.i == 0) / (1000*1e-3)
print ("Frequency of neuron",0,":" , freq,"Hz")
neuron_0_spikes = spikemon.t[spikemon.i == 0]
print ("Inverse of last spike interval:", 1/(neuron_0_spikes[-2] - neuron_0_spikes[-3]))
```

```python
def plotSpikes(ax, neuron_index, M, spikemon, time_start = 0*ms, time_stop = 50*ms, ):
    
    ax.plot(M.t/ms, M.v[neuron_index]/mV, c='black')
    for i,t in enumerate (spikemon.t):
        if spikemon.i[i] == neuron_index:
            ax.axvline(t/ms, c='black')
    
    ax.set(xlabel='Time (ms)', ylabel='v (mV) (neuron {})'.format(neuron_index), ylim = [-80,0], xlim = [time_start/ms, time_stop/ms])

def plotgs(ax, M, spikemon, time_start = 0*ms, time_stop = 50*ms):
    
    ax.plot(M.t/ms, M.gs[0]/nsiemens, label = 'gs in neuron 0')

    init_flag = True
    for i,t in enumerate (spikemon.t):
        if spikemon.i[i] == 1:
            if init_flag:
                ax.axvline(t/ms, ls = '--' , c="red", label = 'spike in neuron 1')
                init_flag = False
            else:
                ax.axvline(t/ms, ls = '--' , c="red")

    ax.set (xlabel = 'Time (ms)', ylabel='gs', xlim = [time_start/ms, time_stop/ms] )
    ax.legend();


start_scope()

IeRm = 25*mV
taum = 20*ms
El = -70*mV
Vth = -54*mV
rm = 10 * Mohm
tau = 10*ms
Es = 0*mV
eqs = '''
dv/dt = (IeRm - (v-El)- gs*(v-Es)*rm)/taum : volt
dgs/dt = -gs/tau : siemens
'''

G = NeuronGroup(2, eqs,threshold='v > Vth', reset='v = El-10*mV', method='exponential_euler')
M = StateMonitor(G, ['v','gs'], record=True)
spikemon = SpikeMonitor(G)

S = Synapses(G, G, on_pre='gs_post += 5 * nsiemens')
S.connect(i=0, j=1)
S.connect(i=1, j=0)
S.delay = '10*ms'
# visualise_connectivity(S)

G.v = [El + 10*mV , El - 10*mV]

run(1*second)

figure()
fig, axs = plt.subplots(2, 2, figsize=(11, 7))
plotSpikes(axs[0,0],0, M, spikemon, time_start = 0*ms, time_stop = 100*ms)
plotSpikes(axs[0,1],0, M, spikemon, time_start = 1000*ms - 100*ms, time_stop = 1000*ms)
plotSpikes(axs[1,0],1, M, spikemon, time_start = 0*ms, time_stop = 100*ms)
plotSpikes(axs[1,1],1, M, spikemon, time_start = 1000*ms - 100*ms, time_stop = 1000*ms)
for ax in axs.flat:
    ax.label_outer()
    
figure()
fig, axs = plt.subplots(1, 2, figsize=(11, 4))
plotgs(axs[0], M, spikemon, time_start = 0*ms, time_stop = 100*ms)
plotgs(axs[1], M, spikemon, time_start = 1000*ms - 100*ms, time_stop = 1000*ms)
# for ax in axs.flat:
#     ax.label_outer()

freq = np.sum (spikemon.i == 0) / (1000*1e-3)
print ("Frequency of neuron",0,":" , freq,"Hz")
neuron_0_spikes = spikemon.t[spikemon.i == 0]
print ("Inverse of last spike interval:", 1/(neuron_0_spikes[-2] - neuron_0_spikes[-3]))
```

همانطور که مشاهده می‌شود، در رفتارهای مهاری دو مسئله مهم مشاهده می‌شود. اول آن که به دلیل کنش مهاری بین دو نورون، فرکانس اسپایک زدن نسبت به حالت تحریکی، کمتر شده است. علت آن است که هر نورون با اسپایک زدن باعث می‌شود اسپایک نورون دیگر به تاخیر بیافتد. همچنین همانطور که مشاهده می‌شود، در 50 میلی ثانیه اول، دو نورون با اختلاف فاز 180 درجه با یکدیگر اسپایک می‌زنند. اما هر چه به انتهای شبیه سازی نزدیک تر می‌شویم، این اختلاف فاز کمتر شده و رفتار دو نورون با هم سنکرون می‌شود. این سنکرون شدن به آن علت است که بهترین پاسخ ممکن برای این دستگاه معادلات دیفرانسیل، با سنکرون شدن دو نورون محقق می‌شود.

```python
import draw_utils
```

```python
def simulateRatePopulation (tau_i = 30 * ms, linear = False, tau_e = 10 * ms, Mee = 1.25, Mei = -1, gamma_e = -10 * Hz, gamma_i =  10 * Hz, Mie = 1, Mii = 0, name = None):
    start_scope()
    
    if not linear:
        eq1 = '''
        vi : Hz (linked)
        dve/dt = (-ve + clip(Mee * ve + Mei * vi - gamma_e, 0 * Hz, 1000 * Hz))/tau_e: Hz
        '''
    
        eq2 = '''
        ve : Hz (linked) 
        dvi/dt = (-vi + clip(Mie * ve + Mii * vi  - gamma_i, 0 * Hz, 1000 * Hz))/tau_i:  Hz
        '''
    else:
        eq1 = '''
        vi : Hz (linked)
        dve/dt = (-ve + Mee * ve + Mei * vi - gamma_e)/tau_e: Hz
        '''
    
        eq2 = '''
        ve : Hz (linked) 
        dvi/dt = (-vi + Mie * ve + Mii * vi  - gamma_i)/tau_i:  Hz
        '''
        

    exc_pop = NeuronGroup(1, eq1, method='exponential_euler')
    exc_pop.ve = 50 * Hz
    inh_pop = NeuronGroup(1, eq2, method='exponential_euler')
    inh_pop.vi = 50 * Hz

    exc_pop.vi = linked_var(inh_pop, 'vi', )
    inh_pop.ve = linked_var(exc_pop, 've')

    Mi = StateMonitor(inh_pop, ['vi'], record=True)
    Me = StateMonitor(exc_pop, ['ve'], record=True)

    run(2000*ms)
    
    def F(X, Y, params):
        
        if not linear:
            return (-X + clip(Mee * X + Mei * Y - gamma_e/Hz, 0, 1000))/(tau_e / ms)
        else:
            return (-X + Mee * X + Mei * Y - gamma_e/Hz)/(tau_e / ms)
        
    def G(X, Y, params):
        
        if not linear:
            return (-Y + clip(Mie * X + Mii * Y  - gamma_i/Hz, 0, 1000))/(tau_i / ms)
        else:
            return (-Y + Mie * X + Mii * Y  - gamma_i/Hz)/(tau_i / ms)

    def f_null1(x, params):
        return -(-x + Mee * x - gamma_e/Hz)/Mei

    def f_null2(x, params):
        return (Mie * x - gamma_i/Hz)/ (1 - Mii)
    def getEigenValues():
        a = (Mee-1)/(tau_e/second)
        b = (Mii-1)/(tau_i/second)
        c = Mie/(tau_i/second)
        d = Mei/(tau_e/second)
        return 1/2*(a + b + np.sqrt ((a-b)**2 + 4*c*d + 0j)) , 1/2*(a + b - np.sqrt ((a-b)**2 + 4*c*d + 0j)),
    
    l1, l2 = getEigenValues()
    print ("   eigen values : {}, {}".format(getEigenValues()[0] , getEigenValues()[1]))
    if np.real(l1) > 0 or np.real(l2) > 0:
        print ("         Unstable fixed point")
    else:
        print ("         Stable fixed point")
    
    if name == None:
        name = "taui,{}".format(tau_i/ms) + (",linear" if linear else ",non-linear")
#     print (name)
    draw_utils.plot_phaseplane(F, G, f_null1, f_null2, x_range=[0.,50.],y_range=[0, 50.], name=name , step = 5)    
#     print (name)
    draw_utils.phaseplane_animation(Me.t/ms, Me.ve[0]/Hz, Mi.vi[0]/Hz, Nullclines=[f_null1, f_null2, None], name=name, fps = 14., animspeed = 28)
    
```

همانطور که در شکل‌های فوق دیده می‌شود، ثابت زمانی 50 میلی‌ثانیه باعث ناپایدار شدن نقطه ثابت می‌شود و مدل خطی ناپایدار خواهد شد. اما در مدل غیر خطی، با پارامترهای مشابه، با توجه به این که نرخ اسپایک زدن نمی‌تواند منفی شود، هر چند نقطه ثابت ناپایدار است اما مسیر حرکت ذره در یک limit cycle محدود خواهد شد. 
    لذا غیر خطی بودن هم مدل را پایدار کرده و هم جلوی منفی شدن نرخ اسپایک را می‌گیرد.
