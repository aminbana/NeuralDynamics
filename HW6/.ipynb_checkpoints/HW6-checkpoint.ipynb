# Q1 - Single layer network

## generating dataset

```python
import numpy as np

N = 100
x_min , x_max = -2, 2

# samples in the range -2 , 2
xs = (x_max - x_min) * np.random.random(size = N) + x_min

assert xs.max() < x_max and x_min < xs.min()
```

```python
# 1-D not function
def NOT(x):        
    out = x.copy()
    out[x >= 0] = 0.
    out[x < 0] = 1.
    return out
y_stars = NOT (xs)

```

## Implementing one epoch

<div style="direction:rtl;line-height:300%;text-align:justify;" align="justify"><font face="B nazanin" size=5>
از توضیحات کتاب
می‌دانیم که رابطه آپدیت پارامترها (
params
)
برای وزن‌های لایه آخر
به صورت زیر تعریف می شود:

```python
# sigmoid function
def sigmoid(x):
    return 1/(1+np.exp (-x))

# computing weight updates given a single data pair and network parameters
def get_weight_update_for_one_data(x,y_star , eta, params):
    
    y,_,_ = get_network_output (x, params)
    
    x_new = [x, -1]
    
    delta_params = []
    for i,p in enumerate (params):
        delta_params.append (- eta * (y - y_star) * y * (1 - y) * x_new[i])
        
    return delta_params

# computing network output given input data and params (both in batch mode and online mode)
def get_network_output (x, params):
    w, theta = params
    return sigmoid (w*x - theta), _ , _ # two extra parameters are added for consistency with 2 layer perceptron

# computing total network error given the dataset and the network parameters
def get_total_error(xs, y_stars, params):
    
    y, _ , _ = get_network_output(xs, params)
    return ((y - y_stars) ** 2).sum()

# performing online learning on all datas for one epoch and updating params
def train_one_epoch(xs, y_stars, eta, params_init, ):
    params = params_init.copy()

    for x, y_star in zip (xs , y_stars):
        error = get_total_error(xs, y_stars, params)

        delta_params = get_weight_update_for_one_data(x , y_star, eta , params)

        for i, delta_p in enumerate (delta_params):
            params[i] = params[i] + delta_p
        
    return error, params


eta = 1
w0 = theta0 = 3
params_init = [w0, theta0]

error, params = train_one_epoch(xs, y_stars, eta, params_init)

print ("Error after one epoch is : " , error)
```

```python
import matplotlib.pyplot as plt

# plotting the target funtion and the estimated function
def plot_in_range(params):
    
    xs = np.linspace(start = x_min , stop = x_max, num=1000)
    ys , _ , _ = get_network_output(xs, params)
    y_stars = NOT (xs)
    plt.plot (xs, y_stars, label = 'ideal')
    plt.plot (xs, ys, label = 'network output')
    
    plt.legend()
    plt.show()

plot_in_range (params)
```

## Training for 100 epochs

```python

# Training for 'epochs' iterations given the initial parameters
def train_multiple_epochs (eta, epochs, params_init):

    params_history = [[] for t in params_init]
    
    errors = []
    params = params_init.copy()
    
    for epoch in range (epochs):
        for t in range(len (params_history)):
            params_history[t].append (params[t])
            
        error, params = train_one_epoch(xs, y_stars, eta, params)
        errors.append (error)
        
    
    for t in range(len (params_history)):
        params_history[t].append (params[t])
    
    final_params = [ps[-1] for ps in params_history]
    
    return errors, params_history, final_params

# plotting the history of wieghts and error during training
def plot_epochs (epochs, errors, params_history, params_names):
    for parameter_history, parameter_name in zip (params_history, params_names):
        plt.plot (parameter_history , label = parameter_name)
        plt.title(parameter_name)
        plt.legend()
        plt.show ()
    
    plt.plot (errors , label = "errors")
    plt.title("Error history")
    plt.legend()
    plt.show ()
```

```python
epochs = 1000
eta = 1
w0 = theta0 = 3
params_init = [w0, theta0]

errors, params_history, final_params = train_multiple_epochs (eta, epochs, params_init)

plot_epochs(epochs, errors, params_history = params_history, params_names = ['W' , 'theta0'])
```

# Q2 - Double input

<div style="direction:rtl;line-height:300%;text-align:justify;" align="justify"><font face="B nazanin" size=5>
در این بخش به پرسش 2 پاسخ داده شده است. توجه کنید که کدهای این قسمت نمی‌توانند مستقل از قسمت‌های فوق اجرا شوند چرا که برخی توابعی که در این بخش استفاده شده است را در بالا تعریف کرده‌ایم.
بنابراین توصیه می‌شود کل نوت بوک را به ترتیب اجرا کنید.

</font></div>

## generating dataset

```python
import numpy as np

N = 100
x_min , x_max = -2, 2

# samples in the range -2 , 2
xs = (x_max - x_min) * np.random.random(size = [N,2]) + x_min

assert xs.max() < x_max and x_min < xs.min()
```

```python
def f1(x):
    out = np.zeros(x.shape[0])
    out[x[:,0] >= 0] = 1.
    out[x[:,0] <  0] = 0.
    
    return out

y_stars = f1 (xs)

```

## train loop

```python
def sigmoid(x):
    return 1/(1+np.exp (-x))

def get_weight_update_for_one_data(x, y_star, eta, params):
    
    y , _ , _ = get_network_output (x, params)
    
    x_new = [x[0], x[1] , -1]
    
    delta_params = []
    for i,p in enumerate (params):
        delta_params.append (- eta * (y - y_star) * y * (1 - y) * x_new[i])
    
    return delta_params

def get_network_output (x, params):
    w1, w2, theta = params
    if x.ndim == 2: #batch mode calculations
        return sigmoid (w1*x[:,0] + w2*x[:,1] - theta) , _ , _
    else:
        return sigmoid (w1*x[0] + w2*x[1] - theta) , _ , _


eta = 1
w01 = w02 = theta0 = 3
params_init = [w01, w02, theta0]
error, params = train_one_epoch(xs, y_stars, eta, params_init)

print ("Error after one epoch is : " , error)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

def plot_in_range(params):
     
    x1 = np.linspace(x_min, x_max, 50)
    x2 = np.linspace(x_min, x_max, 50)

    X1, X2 = np.meshgrid(x1, x2)

    Z = X1.copy()
    for i in range (len(x1)):
        for j in range (len (x2)):
            Z[i,j] = f1(np.array ([x1[i],x2[j]])[None,:])
    
    ax = plt.axes(projection='3d')
    ax.plot_surface(X1, X2, Z, rstride=1, cstride=1,cmap='viridis', edgecolor='none')
    
    ax.set_title ("Target Function")
    ax.set_xlabel ("x2")
    ax.set_ylabel ("x1")
    ax.set_zlabel ("z")
    
    plt.show()
    
    Z = X1.copy()
    for i in range (len(x1)):
        for j in range (len (x2)):
            Z[i,j], _ , _ = get_network_output(np.array ([x1[i],x2[j]]), params)
    
    ax = plt.axes(projection='3d')
    ax.plot_surface(X1, X2, Z, rstride=1, cstride=1,cmap='viridis', edgecolor='none')
    ax.set_title ("Network Function")
    ax.set_xlabel ("x2")
    ax.set_ylabel ("x1")
    ax.set_zlabel ("z")
    
    plt.show()
    

# plot_in_range (params)
```

```python
epochs = 1000
eta = 1
w01 = w02 = theta0 = 3
params_init = [w01, w02, theta0]

errors, params_history, final_params = train_multiple_epochs (eta, epochs, params_init)

plot_epochs(epochs, errors, params_history = params_history, params_names = ['W1' , 'W2' , 'theta0'])
```

## AND function

```python
def AND(x):
    out = np.zeros(x.shape[0])
    out[(x[:,0] >= 0) & (x[:,1] >= 0)] = 1.
    out[(x[:,0] < 0) | (x[:,1] < 0)] = 0.
    
    return out

f1 = AND
y_stars = f1 (xs)
```

```python
epochs = 1000
eta = 1
w01 = w02 = theta0 = 3
params_init = [w01, w02, theta0]

errors, params_history, final_params = train_multiple_epochs (eta, epochs, params_init)

plot_epochs(epochs, errors, params_history = params_history, params_names = ['W1' , 'W2' , 'theta0'])

plot_in_range (final_params)
```

## XOR function

```python
def XOR(x):
    out = np.zeros(x.shape[0])
    out[x[:,0]*x[:,1] <= 0] = 1.
    out[x[:,0]*x[:,1] > 0] = 0.
    
    return out

f1 = XOR
y_stars = f1 (xs)
```

```python
epochs = 1000
eta = 1
w01 = w02 = theta0 = -10
params_init = [w01, w02, theta0]

errors, params_history, final_params = train_multiple_epochs (eta, epochs, params_init)

plot_epochs(epochs, errors, params_history = params_history, params_names = ['W1' , 'W2' , 'theta0'])

plot_in_range (final_params)
```

# Q3 double layer perceptron

## dataset generation

```python
def f1(x):
    out = np.zeros(x.shape[0])
    out[x[:,0] >= 0] = 1.
    out[x[:,0] <  0] = 0.
    
    return out

y_stars = f1 (xs)
```

## train loop

```python
def sigmoid(x):
    return 1/(1+np.exp (-x))

def get_weight_update_for_one_data(x,y_star , eta , params):
    
    y,z1,z2 = get_network_output (x, params)
    
    wp1, wp2, w11, w12, w21, w22, theta1, theta2, theta3 = params
    
    delta_wp1 =     - eta * (y - y_star) * y * (1 - y) * z1
    delta_wp2 =     - eta * (y - y_star) * y * (1 - y) * z2
    delta_theta3 = - eta * (y - y_star)  * y * (1 - y) * (-1)
    
    delta_w11 =     - eta * (y - y_star) * y * (1 - y) * wp1 * z1 * (1-z1) * x[0]
    delta_w12 =     - eta * (y - y_star) * y * (1 - y) * wp1 * z1 * (1-z1) * x[1]
    delta_theta1 = - eta * (y - y_star) * y * (1 - y) * wp1 * z1 * (1-z1) * (-1)

    delta_w21 =     - eta * (y - y_star) * y * (1 - y) * wp2 * z2 * (1-z2) * x[0]
    delta_w22 =     - eta * (y - y_star) * y * (1 - y) * wp2 * z2 * (1-z2) * x[1]
    delta_theta2 = - eta * (y - y_star) * y * (1 - y) * wp2 * z2 * (1-z2) * (-1)

    delta_params = [delta_wp1, delta_wp2, delta_w11, delta_w12, delta_w21, delta_w22, delta_theta1, delta_theta2, delta_theta3]
    return delta_params

def get_network_output (x, params):
    wp1, wp2, w11, w12, w21, w22, theta1, theta2, theta3 = params
    if x.ndim == 2:
        z1 = sigmoid (w11*x[:,0] + w12*x[:,1] - theta1)
        z2 = sigmoid (w21*x[:,0] + w22*x[:,1] - theta2)
        return sigmoid (wp1*z1 + wp2*z2 - theta3), z1, z2
    else:
        z1 = sigmoid (w11*x[0] + w12*x[1] - theta1)
        z2 = sigmoid (w21*x[0] + w22*x[1] - theta2)
        return sigmoid (wp1*z1 + wp2*z2 - theta3), z1, z2


epochs = 1000    
eta = 1
wp10 = wp20 = w110 = w120 = w210 = w220 = theta10 = theta20 = theta30  = 3

params_init = [wp10, wp20, w110, w120, w210, w220, theta10, theta20, theta30]
errors, params_history, final_params = train_multiple_epochs (eta, epochs, params_init)

plot_epochs(epochs, errors, params_history = [], params_names = [])
plot_in_range (final_params)
```

## XOR function

```python
def XOR(x):
    out = np.zeros(x.shape[0])
    out[x[:,0]*x[:,1] <= 0] = 1.
    out[x[:,0]*x[:,1] > 0] = 0.
    
    return out

f1 = XOR
y_stars = f1 (xs)
```
